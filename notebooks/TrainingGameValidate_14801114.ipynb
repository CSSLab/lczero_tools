## Verify Training Game #14801114 per forum message

```python
from lcztools.testing import TarTrainingFile
tf = TarTrainingFile('/Volumes/SeagateExternal/leela_data/training_raw/games14800000.tar.gz')

game = tf.read_game('14801114')
```

```python
print(game.get_pgn())
```

### Lichess link: https://lichess.org/u34wHpNu

```python
import chess
def showprobs(p):
    for k, v in p.items():
        print(f"   {k}: {v:0.5f}")

board = game.get_pc_board()
probs = []
for record in game.records:
    probs.append(record.get_probabilities())


b = chess.Board()
for p, m in zip(probs, board.move_stack):
    display(b)
    showprobs(p)
    san = b.san(m)
    uci = m.uci()
    print(f"Choose: {san:5} {uci:5} [{p[uci.rstrip('n')]:0.5f}]")
    print()
    b.push(m)
display(b)
    
    
```
